```python
# imports
import numpy as np
import matplotlib.pyplot as plt
import mesaPlot as mp
import pandas as pd


# data files
data_file1 = 'part_1' # part one uses 0.02 metalicity, 20 msol
```

```python
colors = [
        'steelblue',  # steel blue
        'firebrick',  # firetruck
        '#008080',  # Teal
        'k',  # black
        'magenta', # magenta
        'green',  # Dark Blue
        '#FFFF00',  # Yellow
        '#800080',  # Purple
        '#40E0D0',  # Turquoise
        '#A52A2A'   # Brown
    ]
```

```python
# load data
data = mp.MESA() #this creates an empty data container 
data.loadHistory(data_file1)
print(data.hist.keys()) # Lists available data values
data = {'center_h1': data.hist.center_h1,
        'center_he4': data.hist.center_he4, 
        'center_c12': data.hist.center_c12,
        'center_o16': data.hist.center_o16,
        'center_n14': data.hist.center_n14,
        'star_age':  data.hist.star_age} # choose what data parts we want
data_df = pd.DataFrame(data)
```

```python
data_df
```

```python
# Plot star age vs center_he4
non_zero_index = np.nonzero(data_df['center_h1'].values - data_df['center_h1'][0])
t0_main = data_df['star_age'][non_zero_index[0][0]]

plt.figure(figsize=(8,6), ) #dpi=300)
plt.plot(data_df['star_age'] - t0_main, data_df['center_h1'], label=f'H', color=colors[0], linestyle='--')
plt.plot(data_df['star_age'] - t0_main, data_df['center_he4'], label='He', color=colors[1], linestyle='-.')

plt.xlabel('Star Age (Years)', size=15)
plt.ylabel('Mass Fraction', size=15)
plt.title('Fraction of H & He Vs Star Age', size =15)
plt.xscale('log')
#plt.xlim(0, np.max(data_df['star_age']) - t0_main)
#plt.yscale('log')
plt.legend()
plt.show()
```

```python
# Plot star age vs center_he4
non_zero_index = np.nonzero(data_df['center_h1'].values - data_df['center_h1'][0])
t0_main = data_df['star_age'][non_zero_index[0][0]]

plt.figure(figsize=(8,6),) #)dpi=300)
#center_n14
plt.plot(data_df['star_age'] - t0_main, data_df['center_c12'], label=f'center_c12', color=colors[2], linestyle='--')
plt.plot(data_df['star_age'] - t0_main, data_df['center_n14'], label='center_n14', color=colors[3], linestyle='-.')
plt.plot(data_df['star_age'] - t0_main, data_df['center_o16'], label='center_o16', color=colors[4], linestyle=':')

plt.xlabel('Star Age (Years)', size=15)
plt.ylabel('Mass Fraction', size=15)
plt.title('Fraction Vs Star Age', size =15)
plt.xscale('log')
plt.yscale('log')
#plt.xlim(0, np.max(data_df['star_age']) - t0_main)
#plt.yscale('log')
plt.legend()
plt.show()
```
