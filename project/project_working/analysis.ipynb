```python
# load data

data_part_2_list = []
for i, data_file in enumerate(data_files_part_2):

        data = mp.MESA() #this creates an empty data container 
        data.loadHistory(data_files2  + data_file)
        print(data.hist.keys()) # Lists available data values
        data = {'center_h1': data.hist.center_h1,
        'center_he4': data.hist.center_he4, 
        'center_c12': data.hist.center_c12,
        'center_o16': data.hist.center_o16,
        'star_mass': data.hist.star_mass,
        'log_L': data.hist.log_L,
        'log_R': data.hist.log_R,
        'log_Teff': data.hist.log_Teff,
        'star_age':  data.hist.star_age} # choose what data parts we want
        data_df = pd.DataFrame(data)
        data_part_2_list.append(data_df)
```

### plot of HR diagram, including pre-main sequence

```python
# plot HR diagram
# 
plt.figure(figsize=(8, 6))
for i, data_file in enumerate(data_files_part_2):
    data_df = data_part_2_list[i]
    log_Teff = data_df['log_Teff']
    log_L = data_df['log_L']
    log_R = data_df['log_R']
    

    #plt.plot(log_Teff, log_L, c=log_R, cmap='viridis',  alpha=0.5, label=data_file)
    plt.plot(log_Teff, log_L, alpha=0.5, label=data_file, linewidth=3, )#)color = colors[i])

# lines of constant R:
#L = 4Ï€ÏƒR2T

# what units is R in? cm? 

R = np.logspace(0.01, 3, 5)
T_arr = np.logspace(3.6, 4.4, 100)
for i in range(len(R)):
    print(i)
    sigma = 5.6704e-5 # cgs
    L = 4*np.pi * sigma *(R[i]*6.9599e10)**2 * T_arr**4
    L_sun = 3.846e33  # cgs
    plt.plot(np.log10(T_arr), np.log10(L/L_sun), color='k', alpha=0.5, linestyle='--',)# label=i) # annotate
    # Annotate the lines with the corresponding R values
    plt.text(np.log10(T_arr[10]), np.log10(L[1] / L_sun), f'R = {float(R[i]):.1f}', ha='left', va='center', fontsize=10, color='k')




#plt.colorbar(label='log_R')
plt.legend()
plt.xlabel("Log(T_eff)", size=15)
plt.ylabel('"log(L)', size=15)
plt.title('HR Diagram', size=15)
plt.gca().invert_xaxis()
plt.ylim(-1, 6)
```
